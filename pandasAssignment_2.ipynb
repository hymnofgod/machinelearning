Group by and Aggregating

```python
import pandas as pd
df=pd.read_csv("drive/My Drive/Colab Notebooks/College/Flavors.csv")
df
```

```python
#Grouping
group_by_frame=df.groupby("Base Flavor")
```

```python
#sum
group_by_frame.sum()
```

```python
#count
group_by_frame.count()
```

```python
#mean
group_by_frame.mean("/content/drive/MyDrive/Colab Notebook/Colleg/Flavours.csv")
```

```python
#describe
group_by_frame.describe()
```

```python
df.groupby(["Base Flavor","Liked"]).agg({'Flavor Rating':['mean','max','count','sum']})
```

```python
df.groupby(["Base Flavor","Liked"]).agg({'Texture Rating':['mean','max','count','sum']})
```

```python
df.groupby(["Base Flavor","Liked"]).agg({'Flavor Rating':['mean','max','count','sum'],'Texture Rating':['mean','max','count','sum']})
```

Merge, Join and Concatinate

```python
import pandas as pd
df1=pd.read_csv("drive/My Drive/Colab Notebooks/College/LOTR.csv")
df1
```

```python
import pandas as pd
df2=pd.read_csv("/content/drive/My Drive/Colab Notebooks/College/LOTR2.csv")
df2
```

```python
pd.concat([df1,df2],ignore_index=True)
```

```python
#inner join (i.e intersection)
df1.merge(df2,how='inner',on=['FellowshipID','FirstName'])
```

```python
#outer join (i.e union)
df1.merge(df2,how='outer')
```

```python
#left join(i.e a-b)
df1.merge(df2,how='left')
```

```python
#right join (i.e b-a)
df1.merge(df2,how='right')
```
