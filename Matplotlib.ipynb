```python
# just an example
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 6)
y = np.array([10, 20, 25, 30, 40])

plt.plot(x, y, marker='o', label="Line")
plt.scatter(x, y, color='red', label="Scatter")
plt.bar(x, y, alpha=0.5, label="Bar")
# plt.legend()
plt.show()

```

```python
#Bar chart
categories=['a','b','c']
values=[10,20,30]

# color=['red','green','blue']

plt.bar(categories,values,color=['red','green','blue'])
plt.title("Bar_Chart_Example")
plt.xlabel("Categories")
plt.ylabel("Values")

plt.show()
```

```python
#reading ice cream rating.csv
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/College/Ice Cream Ratings.csv")
data
```

```python
#set data as index
df=pd.DataFrame(data)
df.set_index('Date',inplace=True)
df
```

```python
#scatter plot using texture rating, overall rating
sns.set_style('whitegrid')
sns.scatterplot(data,x='Texture Rating',y='Overall Rating',color="red")
plt.title("scatter plot")
plt.xlable("Texture Rating" )
plt.ylable("Overall Rating" )
plt.show()
```

```python
# hist plot
sns.histplot(data,kde=True)
plt.title("Histogram")
plt.show()
```

```python
#box plot
sns.boxplot(data)
plt.title("Box plot")
plt.show()
```

```python
import pandas as pd
df=pd.read_excel("/content/drive/MyDrive/Colab Notebooks/College/Customer Call List.xlsx")
df
```

```python
#drop duplicate
df=df.drop_duplicates()
df
```

```python
df1 = df.drop("Not_Useful_column", axis=1,inplace=True)

df
```

```python
df
```

```python
#fix issue with Last Name
df["Last_Name"]=df["Last_Name"].str.replace("[\/._]","",regex=True)
df
```

```python
#make phone number format as 3digits-3digits-remaining digits
#split the column adress and create three column named: street_address, state, zip_code
df[['street_address', 'state', 'zip_code']]=df['Address'].str.split(",", expand=True)
df
```

```python

#replacing Y to Yes and N to No
# df.loc[[2, 4, 7, 11, 14, 17], "Paying Customer"]="No", "Yes", "No", "Yes", "No", "Yes"
df["Paying Customer"]=df["Paying Customer"].replace("Y", "Yes")
df["Paying Customer"]=df["Paying Customer"].replace("N", "No")
df
```

```python
#replacing Y to Yes and N to No
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("Y", "Yes")
df["Do_Not_Contact"]=df["Do_Not_Contact"].replace("N", "No")
df
```

```python

df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("NaN", "",)
df
```
