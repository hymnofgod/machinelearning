```python

# Sample DataFrame
data = {
    "Duration": [0, 1, 2, 3, 4],
    "Date": ["2020/12/01'", "2020/12/02'", "2020/12/03'", "2020/12/04'", "2020/12/05'"],
    "Pulse": [110, 117, 103, 109, 117],
    "Maxpulse": [130, 145, 135, 175, 148],
    "Calories": [409.1, 479.0, 340.0, 282.4, 406.0]
}

df = pd.DataFrame(data)

# Removing apostrophe from the Date column
df["Date"] = df["Date"].str.replace("'", "", regex=False)

print(df)
```
